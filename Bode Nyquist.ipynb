```python
#Bode Plot & Phase Margin
#Using control module

import numpy as np
import matplotlib.pyplot as plt
import control as ct

#From Ogata, Example 7-21, p.469 
sys = ct.tf([20,20], np.convolve([1, 5, 0],[1,2,10]))
response = ct.bode_plot([sys])

gm, pm, wcg, wcp = ct.margin(sys)
gmdB = 20*np.log10(gm)
print(gmdB)
```

```python
#Bode Plot & Phase Margin
#Using control module

import numpy as np
import matplotlib.pyplot as plt
import control as ct

#From Ogata, Example 7-20, p.467 
sys1 = ct.tf([10], [1, 6, 5, 0])
response = ct.bode_plot([sys1])

gm, pm, wcg, wcp = ct.margin(sys1)
gmdB = 20*np.log10(gm)
print('gain margin sys1=', gmdB)
print('phase margin sys1=', pm)
print()
sys2 = ct.tf([100], [1, 6, 5, 0])
response = ct.bode_plot([sys2])

gm, pm, wcg, wcp = ct.margin(sys2)
gmdB = 20*np.log10(gm)
print('gain margin sys2=', gmdB)
print('phase margin sys2=', pm)
```

```python
#Nyquist Plot
fig=plt.figure(figsize=(10, 6))
ct.nyquist_plot([sys1, sys2])
```
